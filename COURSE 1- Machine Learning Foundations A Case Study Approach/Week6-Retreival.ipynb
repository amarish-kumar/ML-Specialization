```python
import graphlab

# Limit number of worker processes. This preserves system memory, which prevents hosted notebooks from crashing.
graphlab.set_runtime_config('GRAPHLAB_DEFAULT_NUM_PYLAMBDA_WORKERS', 8)
```

```python
#load CIFAR-10 dataset
image_train = graphlab.SFrame('image_train_data/')
image_test = graphlab.SFrame('image_test_data/')

#is similar to retrieving documents task
```

```python
#traina a nearest-nerighbors model for retrieving images
knn_model = graphlab.nearest_neighbors.create(image_train, features= ['deep_features'], label = 'id')
```

```python
#take the model to retrieve similar images
graphlab.canvas.set_target('ipynb')
cat = image_train[18:19]
cat['image'].show()
knn_model.query(cat)
```

```python
#a simple function to view the nearest neighbors to save typing:
def get_images_from_ids(query_result):
    return image_train.filter_by(query_result['reference_label'], 'id')

cat_neighbors = get_images_from_ids(knn_model.query(cat))
print cat_neighbors
```

```python
#lambda function to find and show images
show_neighbors = lambda i: get_images_from_ids(knn_model.query(image_train[i:i+1]))['image'].show()
```

```python
#question4

image_test_automobile = image_test.filter_by(['automobile'],'label')
image_test_cat = image_test.filter_by(['cat'],'label')
image_test_dog = image_test.filter_by(['dog'],'label')
image_test_bird = image_test.filter_by(['bird'],'label')

image_test_cat_model = graphlab.nearest_neighbors.create(image_test_cat, features=['deep_features'], label='id')
image_test_dog_model = graphlab.nearest_neighbors.create(image_test_dog, features=['deep_features'], label='id')
image_test_bird_model = graphlab.nearest_neighbors.create(image_test_bird, features=['deep_features'], label='id')
image_test_automobile_model = graphlab.nearest_neighbors.create(image_test_automobile, features=['deep_features'], label='id')
```

```python
dog_cat_neighbors = image_test_cat_model.query(image_test_dog, k=1)
dog_dog_neighbors = image_test_dog_model.query(image_test_dog, k=1)
dog_bird_neighbors = image_test_bird_model.query(image_test_dog, k=1)
dog_automobile_neighbors = image_test_automobile_model.query(image_test_dog, k=1)
```

```python
dog_distances = graphlab.SFrame({'dog_automobile': dog_automobile_neighbors['distance'],
                              'dog_bird': dog_bird_neighbors['distance'],
                              'dog_cat': dog_cat_neighbors['distance'],
                              'dog_dog': dog_dog_neighbors['distance']
                             })
```

```python
def is_dog_correct(row):  
    if row['dog_dog'] <= min(row.values()):     
        return 1    
    else:        
        return 0

dog_distances.apply(is_dog_correct).sum()
```

```python

cat_distances = graphlab.SFrame({'cat_automobile': automobile_model.query(image_test_cat, k=1)['distance'],
                                 'cat_bird': bird_model.query(image_test_cat, k=1)['distance'],
                                 'cat_cat': cat_model.query(image_test_cat, k=1)['distance'],
                                 'cat_dog': dog_model.query(image_test_cat, k=1)['distance'],
                                })


def is_cat_correct(row):  
    if row['cat_cat'] <= min(row.values()):     
        return 1    
    else:        
        return 0
    

cat_distances.apply(is_cat_correct).sum()
```
